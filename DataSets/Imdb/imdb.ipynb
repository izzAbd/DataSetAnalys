```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("users.data", sep='\t', names = ['user_id','item_id','rating','timestamp'])
```

```python
data
```

```python
data2 = pd.read_csv("movie_id_titles.csv")
```

```python
data2
```

```python
data = pd.merge(data,data2,on="item_id")
```

```python
data
```

```python
tablo = data.pivot_table(index="user_id",columns="title",values="rating")
```

```python
tablo
```

```python
tablo.head()
```

```python
starwars = tablo['2001: A Space Odyssey (1968)']
```

```python
starwars.head()
```

```python
korelasiya = tablo.corrwith(starwars)
```

```python
korelasiya
```

```python
kore = pd.DataFrame(korelasiya,columns=["Correlation"])
kore.dropna(inplace=True)
```

```python
kore
```

```python
kore.head(10)
```

```python
kore.tail(10)
```

```python
kore.sort_values("Correlation",ascending=False)
```

```python
kore.head(15)
```

```python
ratings = pd.DataFrame(data.groupby("title")["rating"].mean())
ratings.sort_values("rating",ascending=False).head()
```

```python
ratings ["point_count"] = pd.DataFrame(data.groupby("title")['rating'].count())
```

```python
ratings.head()
```

```python
ratings.sort_values("point_count",ascending=False).head()
```

```python
kore.head()
```

```python
kore = kore.join(ratings['point_count'])
```

```python
kore[kore['point_count'] > 100].sort_values("Correlation",ascending=False).head()
```
